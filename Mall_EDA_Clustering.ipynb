```python
import pandas as pd

# Assuming you have a DataFrame named 'df'
def categorize_columns(df):
    categorical_cols = []
    numerical_cols = []
    mixed_cols = []

    for col in df.columns:
        if df[col].dtype == 'object':
            if df[col].nunique() == len(df):
                mixed_cols.append(col)
            else:
                categorical_cols.append(col)
        elif pd.api.types.is_numeric_dtype(df[col]):
            numerical_cols.append(col)
        else:
            mixed_cols.append(col)

    print("Categorical Columns:")
    print(categorical_cols)
    
    print("\nNumerical Columns:")
    print(numerical_cols)
    
    print("\nMixed Columns:")
    print(mixed_cols)

# Call the function with your DataFrame
categorize_columns(df)
```

## Segemation using Various Clustering Techniques

```python
numeric=df[['Age','Annual Income (k$)', 'Spending Score (1-100)']]
numeric.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Define the range of n_clusters to test
n_clusters_range = [2, 3, 4, 5, 6, 7, 8, 9]

# Initialize lists to store inertia and silhouette scores
cluster_inertia = []
silhouette_scores = []

# Iterate over the range of n_clusters
for n_clusters in n_clusters_range:
    # Create a KMeans model with the current n_clusters
    kmeans = KMeans(n_clusters=n_clusters, n_init=25)
    
    # Fit the model to the data
    kmeans.fit(numeric)
    
    # Get the inertia (within-cluster sum of squares)
    inertia = kmeans.inertia_
    
    # Calculate the silhouette score
    labels = kmeans.labels_
    score = silhouette_score(numeric, labels)
    
    # Append the inertia and silhouette score to the respective lists
    cluster_inertia.append(inertia)
    silhouette_scores.append(score)

# Plot the Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(n_clusters_range, cluster_inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()

```

```python
# Plot the Silhouette Score
plt.figure(figsize=(8, 6))
plt.plot(n_clusters_range, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score')
plt.show()
```

```python
# here the silhoute score is high for the 6 clusters. Silhoute score is would indicate best options for the clusters
# we can take k as 6 clusters
```

```python
# K-Means clustering with 6 clusters
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
kmeans = KMeans(n_clusters=6, random_state=42)
df['KMeans_Cluster'] = kmeans.fit_predict(numeric)
```

```python
# Visualize the clusters
sns.pairplot(df, hue='KMeans_Cluster', palette='Set1', diag_kind='hist')
plt.suptitle('K-Means Clustering')
plt.show()
```

```python
# Get the cluster labels
k_labels = kmeans.labels_

# Create a DataFrame with the cluster labels
cluster_df = pd.DataFrame({'KMeans_Cluster': k_labels})

# Display the number of clusters and their sizes
print("Number of Clusters:", kmeans.n_clusters)
print("Cluster Sizes:")
print(cluster_df['KMeans_Cluster'].value_counts())
```

```python
# DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=3)
df['DBSCAN_Cluster'] = dbscan.fit_predict(numeric)
```

```python
# Remove the 'KMeans_Cluster' column from the DataFrame
df = df.drop('KMeans_Cluster', axis=1)

# Create the pair plot for the 'DBSCAN_Cluster' column
sns.pairplot(df, hue='DBSCAN_Cluster', palette='Set1', diag_kind='hist')
plt.suptitle('DBSCAN Clustering')
plt.show()
```

```python
dbscan_labels = dbscan.labels_
# Add the DBSCAN cluster labels to the DataFrame
cluster_df['DBSCAN_Cluster'] = dbscan_labels

# Display the number of clusters and their sizes
print("Number of Clusters:", kmeans.n_clusters)
print("Cluster Sizes:")
print(cluster_df['KMeans_Cluster'].value_counts())
print(cluster_df['DBSCAN_Cluster'].value_counts())
```

```python
# Agglomerative clustering
agg = AgglomerativeClustering(n_clusters=6)
df['Agglomerative_Cluster'] = agg.fit_predict(numeric)
```

```python
# Remove the 'KMeans_Cluster' column from the DataFrame
df = df.drop('DBSCAN_Cluster', axis=1)
sns.pairplot(df, hue='Agglomerative_Cluster', palette='Set1', diag_kind='hist')
plt.suptitle('Agglomerative Clustering')
plt.show()
```

```python
# Get the cluster labels for Agglomerative clustering
agg_labels = agg.labels_
# Add the Agglomerative cluster labels to the DataFrame
cluster_df['Agglomerative_Cluster'] = agg_labels
```

```python
# Display the number of clusters and their sizes for all three clustering algorithms
print("Number of Clusters - KMeans:", kmeans.n_clusters)
print("Number of Clusters - DBSCAN:", len(set(dbscan_labels)))
print("Number of Clusters - Agglomerative:", len(set(agg_labels)))

print("\nCluster Sizes - KMeans:")
print(cluster_df['KMeans_Cluster'].value_counts())

print("\nCluster Sizes - DBSCAN:")
print(cluster_df['DBSCAN_Cluster'].value_counts())

print("\nCluster Sizes - Agglomerative:")
print(cluster_df['Agglomerative_Cluster'].value_counts())
```

```python
# Calculate the silhouette score for KMeans labels
kmeans_silhouette = silhouette_score(numeric, kmeans.labels_)
print("Silhouette Score for KMeans labels:", kmeans_silhouette)

# the number of labels is 1, which is not valid for the silhouette score calculation.
#The silhouette score requires at least two distinct clusters to calculate the silhouette coefficient.
#so we cant calculate for dbscan
# Calculate the silhouette score for Agglomerative labels
agglomerative_silhouette = silhouette_score(numeric, agg.labels_)
print("Silhouette Score for Agglomerative labels:", agglomerative_silhouette)
```
